#**Import all required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Load dataset**

```python
df = pd.read_csv("/content/ISRO Satellite Dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

#**Select important columns**

```python
df = df[["Current Official Name of Satellite", "Operator/Owner", "Users", "Purpose", "Class of Orbit", "Longitude of GEO (degrees)", "Perigee (km)", "Apogee (km)", "Period (minutes)", "Launch Mass (kg.)", "Power (watts)", "Date of Launch", "Expected Lifetime (yrs.)", "Country of Contractor", "Launch Vehicle"]]
```

# **Drop some columns**

```python
df_isro = df.drop(["Users", "Country of Contractor",], axis=1)
```

#**Line plot for Apogee by date of launch**

```python
plt.plot(df_isro["Date of Launch"], df_isro["Apogee (km)"])
plt.title("Apogee by Date of launch")
plt.xlabel("Date of launch")
plt.ylabel("Apogee")
plt.show()
```

#**Line plot for Perigee by date of launch**

```python
plt.plot(df_isro["Date of Launch"], df_isro["Perigee (km)"])
plt.title("Perigee by Date of launch")
plt.xlabel("Date of launch")
plt.ylabel("Perigee")
plt.show()
```

#**Bar plot**

```python
plt.bar(df_isro['Class of Orbit'], df_isro['Expected Lifetime (yrs.)'])
plt.title("Expected Lifetime (yrs) by Class of Orbit")
plt.xlabel("Class of Orbit")
plt.ylabel("Expected Lifetime")
plt.show()
```

```python
plt.bar(df_isro['Class of Orbit'], df_isro['Apogee (km)'])
plt.title("Apogee (km) by Class of Orbit")
plt.xlabel("Class of Orbit")
plt.ylabel("Apogee (km)")
plt.show()
```

```python
plt.bar(df_isro['Class of Orbit'], df_isro['Perigee (km)'])
plt.title("Perigee (km) by Class of Orbit")
plt.xlabel("Class of Orbit")
plt.ylabel("Perigee (km)")
plt.show()
```

#**Horizontal bar plot**

#**Histogram plot**

```python
plt.hist(df_isro["Apogee (km)"], bins=10, color='skyblue', edgecolor='black')
plt.title("Frequency by Apogee")
plt.xlabel("Apogee")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df_isro["Perigee (km)"], bins=10, color='skyblue', edgecolor='black')
plt.title("Frequency by Perigee")
plt.xlabel("Perigee")
plt.ylabel("Frequency")
plt.show()
```

#**Scatter plot**

```python
plt.scatter(df_isro["Apogee (km)"], df_isro["Perigee (km)"], color='green')
plt.title("Apogee vs Perigee")
plt.xlabel("Apogee")
plt.ylabel("Perigee")
plt.show()
```

```python
plt.boxplot(df_isro['Perigee (km)'])
plt.title("Perigee")
plt.ylabel("Values")
plt.show()
```

```python
plt.boxplot(df_isro['Apogee (km)'])
plt.title("Apogee")
plt.ylabel("Values")
plt.show()
```

```python
plt.violinplot(df_isro['Period (minutes)'])
plt.title("Violin Plot")
plt.ylabel("Values")
plt.show()
```
